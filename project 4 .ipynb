## univariate analysis 
Univariate analysis is a statistical method used to describe and understand the characteristics of a single variable. It involves analyzing one variable at a time to understand its distribution, central tendency, dispersion, and shape of the data. This analysis helps in summarizing and visualizing data to extract insights and patterns.

## bivariate analysis 
Bivariate analysis is a form of quantitative (statistical) analysis that involves the analysis of two variables. It is used to determine the relationship between these variables

```python
# better chance of surivival ? 
val=df.Survived==True
df1=df['Sex'][val]
ans=df['Sex'][val].value_counts()
ans/len(df1)*100

#female had better chance of survival than men 68 % of them survived 
```

```python
# which class survived the most ?
Val=df.Pclass.value_counts()
for i,j in enumerate(Val):
    print(i,j)
Val[3] #important 
```

```python
ams=df['Pclass'][val].value_counts()
ams/len(df[val])*100
#first class survived the most 
```
